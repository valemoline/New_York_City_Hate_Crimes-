# New York City Hate Crime Analysis 

### Project Overview 
This data analysis project is a comparative analysis of different hate crimes between race, gender, and sexual orientation across the 5 boroughs of New York City in 2019-2023. By analyzing the various aspects of the data, we seek to identify trends in current events and their relation to hate crimes to provide insights for the NYPD to help predict which group will have an increase in hate crimes for specific world events. 

### Data Source 
the primary dataset used for this project was: https://data.cityofnewyork.us/Public-Safety/NYPD-Hate-Crimes/bqiq-cu78


### Data Cleaning 
For the initial data-cleaning stage we performed the following tasks: 
1. Inspection of the data
2. Remove columns that do not fit into our outline
3. Convert missing values into 0's

```python
#Data Processing 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
crimes = pd.read_csv("NYPD_Hate_Crimes_2023.csv")
crimes.head()
```

```python
# Cleaning the data 
#dropping columns that are unnessacry towards answering our questions 
crimes.drop(['Record Create Date', 'Complaint Precinct Code', 'Offense Description','PD Code Description'], axis=1)
```

```python
#Filling in missing data 
crimes
crimes.fillna(0)
```

### Percentage and Count of Bias Motive over the years 2019-2023

```python
countcrime=crimes['Bias Motive Description'].value_counts()
percrime=crimes['Bias Motive Description'].value_counts(normalize=True).mul(100).round(1)
pd.concat([countcrime,percrime], axis =1, keys=['count', 'percentage'])
```

This chart displays the total count and percenatage of the hate crimes commited over the past 4 years, This will make it easier for us to look back to for our comparative analysis.

## 1. Which hate crime had the most arrests made 

```python
#Creating a bar graph 
ax = crimes.groupby(['Bias Motive Description'])['Arrest Id'].count().sort_values(ascending=False).head(10).plot(kind='bar',
                                                                          figsize=(14,8),
                                                                           fontsize=(18))
                                                                           
                                                                           
ax.set_title("Top 10 Arrest made Based on Hate Crime Biased Group 2019-2023", fontdict={'fontsize':24})
ax.set_xlabel("Bias Group",fontdict={'fontsize':20})                                                                       
ax.set_ylabel("Frequency",fontdict={'fontsize':20})


```

## 2. What is the number of crimes per year for Anti-Black, Anti-Asian, Anti-Jewish, and Anti-Hispanic? 3. Did Covid-19 and the BLM protests or other world impacts, have an influence on these hate crimes?

### Anti-Black 

```python
condition=crimes['Bias Motive Description'].str.contains('ANTI-BLACK')
Anti_Black_per_month=crimes[condition]
Anti_Black_per_month.groupby('Complaint Year Number')['Bias Motive Description'].count()
graph2=Anti_Black_per_month.groupby('Complaint Year Number')['Bias Motive Description'].count()
ax = graph2.plot(kind='line',
                figsize=(10,5),
                title="Anti Black Crimes per Year")
ax.set_xlabel("Year")
ax.set_ylabel("Frequency")
```

After the murder of George Floyd in 2020, the BLM Protests, and riots started to take to the streets. Here we can see the impact as Anti-Black hate crimes start to take a rise, as police brutality also started to take place to disrupt the protests. After through research on events that have taken place in 2021 and 2020 theres seems to be no correlation as to why Anti-Black crimes continue to increase. Though it has seems that starting now in 2023 anti-black crimes has started to decrease.  

### Anti-Asian 

```python
condition=crimes['Bias Motive Description'].str.contains('ANTI-ASIAN')
Anti_Black_per_month=crimes[condition]
Anti_Black_per_month.groupby('Complaint Year Number')['Bias Motive Description'].count()
graph2=Anti_Black_per_month.groupby('Complaint Year Number')['Bias Motive Description'].count()
ax = graph2.plot(kind='line',
                figsize=(10,5),
                title="Anti Asian Crimes per Year")
ax.set_xlabel("Year")
ax.set_ylabel("Frequency")
```

In 2019 was the start of the pandemic COVID-19 which orginated in China. This led to many Americans believing that all Asian's were the cause to this pandemic prompting to an increase in hate crimes toawards them. This analysis perfeclty reflects that as we see the a slow increase starting in 2019 and a steadily rise till 2021 once the pandemic was decleared over by Goverment officals, which leads towards a decrease in crimes. 

### Anti-Jewish

```python
condition=crimes['Bias Motive Description'].str.contains('ANTI-JEWISH')
Anti_Black_per_month=crimes[condition]
Anti_Black_per_month.groupby('Complaint Year Number')['Bias Motive Description'].count()
graph2=Anti_Black_per_month.groupby('Complaint Year Number')['Bias Motive Description'].count()
ax = graph2.plot(kind='line',
                figsize=(10,5),
                title="Anti Jewish Crimes per Year")
ax.set_xlabel("Year")
ax.set_ylabel("Frequency")
```

It seems that Anti-Jewish has in influx in hate crimes most than any other bias crime. In 2020 it seems to have drop unlike Anti-Black, and Anti-Asian crimes were the main focus as most world events were surrounding them. It slowly starts to increase more when social media, and news articles start to post about the conflict in Palenstine and Isreal starting in 2021. Based on what we have seen with Anti-Black crimes, and Anti-Asian crimes and news surrounding their bias group, we should start to see another spike in Anti-Jewish crimes.
